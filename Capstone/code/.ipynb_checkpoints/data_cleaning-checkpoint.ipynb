### 3. Data Cleaning

```python
#Filtering the dataframe to only the fields that will be used.
nba_data = nba_data[['gmDate', 'seasonYR','teamAbbr','teamLoc','teamRslt','teamDayOff',
                    'teamTO','teamPF','teamFGA','teamFGM','teamFTA','teamFTM',
                    'teamORB','teamDRB','teamEFG%','teamTREB%','teamTO%','teamFT%']]
```

With the referees removed the games can be compressed to two lines for each game by removing duplicates.

```python
nba_data = nba_data.drop_duplicates()
nba_data = nba_data.reset_index(drop=True)
nba_data['gmDate'] = pd.to_datetime(nba_data['gmDate'])
nba_data.head()
```

```python
nba_data.shape
```

```python
#Converted the teamRsl from Win and Loss to O for Loss and 1 for Win.
#Converted the teamLoc from Home and Away to O for Away and 1 for Home.
nba_data['teamRslt'] = nba_data['teamRslt'].apply(lambda x: 0 if x=='Loss' else 1)
nba_data['teamLoc'] = nba_data['teamLoc'].apply(lambda x: 0 if x=='Away' else 1)
```

```python
#Coverting two of the percentage columns from decimals
nba_data['teamEFG%'] = round((nba_data['teamEFG%'] * 100),4)
nba_data['teamFT%'] = round((nba_data['teamFT%'] * 100),4)
```

```python
#Saving data to a new file
nba_data.to_csv('./data/nba_data.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, accuracy_score, confusion_matrix
import pandas_profiling as pp
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', 9999)
pd.set_option('display.max_rows', 200)
```

```python
nba_data = pd.read_csv('./data/nba_data.csv')
```

```python
#Added month column
nba_data.insert(1, "month", 0)
```

```python
nba_data['gmDate'] = pd.to_datetime(nba_data['gmDate'])
nba_data['month'] = pd.DatetimeIndex(nba_data['gmDate']).month
```

### 1. Moving Average Creation

```python
nba_ma = nba_data[['gmDate','month','seasonYR','teamAbbr','teamLoc','teamRslt','teamDayOff']]
```

```python
teams = nba_ma['teamAbbr'].unique()
```

```python
#Create empty dictionaries to hold moving averages
fga = {}
fgm = {} 
to = {}
orb = {}
drb = {}
fta = {}
ftm = {}
pf = {}
efg = {}
top = {}
treb = {}
ftp = {}


for team in teams:
    if team not in fga:
        fga[team] = []


for team in teams:
    if team not in fgm:
        fgm[team] = []
        

for team in teams:
    if team not in to:
        to[team] = []
        

for team in teams:
    if team not in orb:
        orb[team] = []
        
for team in teams:
    if team not in drb:
        drb[team] = []

for team in teams:
    if team not in fta:
        fta[team] = []
        
for team in teams:
    if team not in ftm:
        ftm[team] = []
        
for team in teams:
    if team not in pf:
        pf[team] = []       
        
for team in teams:
    if team not in efg:
        efg[team] = []
    
for team in teams:
    if team not in top:
        top[team] = []
        
for team in teams:
    if team not in treb:
        treb[team] = []
        
for team in teams:
    if team not in ftp:
        ftp[team] = []    
```

```python
#Populate dictionaries with 5 period moving averages
count = 0
k = 5 

for count in range(0, len(nba_ma.index)):
    team_name = nba_ma.loc[count, "teamAbbr"]
    
    if len(fga[team_name]) == k:
        nba_ma.loc[count,'MAteamFGA'] = np.mean(fga[team_name])
        nba_ma.loc[count,'MAteamFGM'] = np.mean(fgm[team_name])
        nba_ma.loc[count,'MAteamTO'] = np.mean(to[team_name])
        nba_ma.loc[count,'MAteamORB'] = np.mean(orb[team_name])
        nba_ma.loc[count,'MAteamDRB'] = np.mean(drb[team_name])
        nba_ma.loc[count,'MAteamFTA'] = np.mean(fta[team_name])
        nba_ma.loc[count,'MAteamFTM'] = np.mean(ftm[team_name])
        nba_ma.loc[count,'MAteamPF'] = np.mean(pf[team_name])
        nba_ma.loc[count,'MAteamEFG%'] = np.mean(efg[team_name])
        nba_ma.loc[count,'MAteamTO%'] = np.mean(top[team_name])
        nba_ma.loc[count,'MAteamTREB%'] = np.mean(treb[team_name])
        nba_ma.loc[count,'MAteamFT%'] = np.mean(ftp[team_name])


        # delete oldest entry for prediction
        fga[team_name].pop(0)
        fgm[team_name].pop(0)
        to[team_name].pop(0)
        orb[team_name].pop(0)
        drb[team_name].pop(0)
        fta[team_name].pop(0)
        ftm[team_name].pop(0)
        pf[team_name].pop(0)
        efg[team_name].pop(0)
        top[team_name].pop(0)
        treb[team_name].pop(0)
        ftp[team_name].pop(0)

        

    if len(fga[team_name]) < k:
        # write data for first team
        fga[team_name].append(nba_data.loc[count, "teamFGA"])
        fgm[team_name].append(nba_data.loc[count, "teamFGM"])
        to[team_name].append(nba_data.loc[count, "teamTO"])
        orb[team_name].append(nba_data.loc[count, "teamORB"])
        drb[team_name].append(nba_data.loc[count, "teamDRB"])
        fta[team_name].append(nba_data.loc[count, "teamFTA"])
        ftm[team_name].append(nba_data.loc[count, "teamFTM"])
        pf[team_name].append(nba_data.loc[count, "teamPF"])
        efg[team_name].append(nba_data.loc[count, "teamEFG%"])
        top[team_name].append(nba_data.loc[count, "teamTO%"])
        treb[team_name].append(nba_data.loc[count, "teamTREB%"])
        ftp[team_name].append(nba_data.loc[count, "teamFT%"])
```

```python
#Exclude the first month of each season
months = [10,11]
nba_ma = nba_ma[(~nba_ma['month'].isin(months))]
nba_ma = nba_ma.reset_index(drop=True)
```

### 3. Correlations and Analysis

```python
nba_corr = round((nba_ma[['teamDayOff', 'teamLoc','MAteamFGA', 'MAteamFGM', 'MAteamTO', 'MAteamORB',
       'MAteamDRB', 'MAteamFTA', 'MAteamFTM', 'MAteamPF','teamRslt']].corr()),2)
sns.set(style="white")
mask = np.zeros_like(nba_corr.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
plt.subplots(figsize=(20,20))
sns.set(font_scale=1.25)
sns.heatmap(nba_corr, xticklabels=nba_corr.columns.values,
            yticklabels=nba_corr.columns.values,mask=mask, cmap = 'coolwarm', annot= True, square=True)
plt.yticks(rotation=0,fontsize = 15)
plt.xticks(rotation=90, fontsize=15);
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(nba_corr.corr()[['teamRslt']].sort_values('teamRslt',ascending=False),vmin=-1, cmap='coolwarm',
            annot=True, annot_kws={"size": 18})
plt.yticks(size=18)
plt.xticks(size=18);
```

```python
nba_corr = round((nba_ma[['teamDayOff', 'teamLoc','MAteamEFG%','MAteamTO%', 'MAteamTREB%',
                            'MAteamFT%', 'MAteamPF','teamRslt']].corr()),2)
sns.set(style="white")
mask = np.zeros_like(nba_corr.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
plt.subplots(figsize=(20,20))
sns.set(font_scale=1.25)
sns.heatmap(nba_corr, xticklabels=nba_corr.columns.values,
            yticklabels=nba_corr.columns.values,mask=mask, cmap = 'coolwarm', annot= True, square=True)
plt.yticks(rotation=0,fontsize = 15)
plt.xticks(rotation=90, fontsize=15);
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(nba_corr.corr()[['teamRslt']].sort_values('teamRslt',ascending=False),vmin=-1, cmap='coolwarm',
            annot=True, annot_kws={"size": 18})
plt.yticks(size=18)
plt.xticks(size=18);
```

### 4. Feature And Model Creation

```python
features = ['teamDayOff', 'teamLoc','MAteamFGA', 'MAteamFGM', 'MAteamTO', 'MAteamORB',
       'MAteamDRB', 'MAteamFTA', 'MAteamFTM', 'MAteamPF']
```

```python
train = nba_ma.loc[(nba_data['seasonYR'] <= 2015)]
test = nba_ma.loc[(nba_data['seasonYR'] > 2015)]
```

```python
nba_ma.head()
```

```python
X_train = train[features]
X_test = test[features]
y_train = train['teamRslt']
y_test = test['teamRslt']
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
preds = lr.predict(X_test)
```

```python
coefficients = pd.DataFrame(dict(zip(nba_ma[features].columns,lr.coef_[0])),index=[0])
coefficients
```

```python
resid = y_test - preds
```

```python
accuracy_score(y_train,lr.predict(X_train))
```

```python
accuracy_score(y_test,lr.predict(X_test))
```

```python
test['predWin'] = preds
```

```python
r = test.iloc[::2,:]['teamRslt']
p = test.iloc[::2,:]['predWin']
```

```python
cm = confusion_matrix(r,p)
tn, fp, fn, tp = confusion_matrix(r, p).ravel()
```

```python
print('True Negatives: ', tn)
print('False Positives: ', fp)
print('False Negatives: ', fn)
print('True Positives: ', tp)
```

```python
#Classification metrics
accuracy = round((((tp+tn)/(tp+fn+tn+fp))*100),2)
misclassification = round(((1 - (accuracy/100))*100),2)
sensitivy = round(((tp/(tp+fn))*100),2)
specificity = round(((tn/(tn+fp))*100),2)
precision = round(((tp/(tp+fp))*100),2)
print('Accuracy Rate:',accuracy,'%')
print('Misclassification Rate:',misclassification,'%')
print('Sensitivy Rate:',sensitivy,'%')
print('Specificity Rate:',specificity,'%')
print('Precision Rate:',precision,'%')
```

```python
cm_df = pd.DataFrame(cm, columns=['pred negative', 'pred positive'], index=['actual negative', 'actual positive'])
cm_df
```

```python
results = pd.DataFrame(test.groupby(['teamAbbr']).agg('sum')[['teamRslt','predWin']])
results.sort_values(by=['teamRslt'], inplace=True)
results = results.reset_index()
```

```python
results['winDiff'] = results['teamRslt'] - results['predWin']
results.head()
```

```python

plt.figure(figsize=(20,15))
plt.style.use('seaborn-whitegrid')
plt.grid(b=None)
plt.yticks(fontsize=25,color='red')
plt.xticks(fontsize=25,color='red')

clist = [(0, "red"), (0.125, "red"), (0.25, "orange"), (0.5, "green"), 
         (0.7, "green"), (0.75, "blue"), (1, "blue")]
rvb = mcolors.LinearSegmentedColormap.from_list("", clist)
width = 0.5
N = 60
q = np.arange(N).astype(float)
plt.barh(results['teamAbbr'],results['teamRslt'],color=rvb(q/N),align='edge');
plt.barh(results['teamAbbr'],results['winDiff'],align='edge')
plt.title('Actual Wins vs Difference In Predictions (Blue)',fontsize=25,color='red');
```

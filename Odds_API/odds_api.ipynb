```python
#set the sport you want to retrieve
sport_key = 'basketball_nba'
```

```python
#retrieve spreads
spreads_response = requests.get('https://api.the-odds-api.com/v3/odds', params={
    'api_key': api_key,
    'sport': sport_key,
    'region': 'us', # uk | us | eu | au - the region of book makers you want to retrieve
    'mkt': 'spreads', # h2h | spreads | totals - choose the type of line you want to retrieve
    'oddsFormat':'american', #retrieve in american or decimal form
    'dateFormat':'unix' #unix or iso date format
})
spreads_json = json.loads(spreads_response.text)
```

```python
#convert JSON to dataframe
columns=['id','home team','away team','start time','site','home spread','away spread',
        'home odds','away odds'] #set columns
spreads_df = pd.DataFrame(columns = columns)
for i in range(0,len(spreads_json['data'])):
    for k in range(i,len(spreads_json['data'][i]['sites'])):
        spreads_df=spreads_df.append(pd.Series([spreads_json['data'][i]['id'],
                                spreads_json['data'][i]['teams'][0], spreads_json['data'][i]['teams'][1],
                                #turn unix date into usable date format
                               dt.datetime.fromtimestamp(spreads_json['data'][i]['commence_time']).strftime('%Y-%m-%d %H:%M:%S'),
                               spreads_json['data'][i]['sites'][k]['site_nice'],
                              spreads_json['data'][i]['sites'][k]['odds']['spreads']['points'][0], 
                               spreads_json['data'][i]['sites'][k]['odds']['spreads']['points'][1],
                               spreads_json['data'][i]['sites'][k]['odds']['spreads']['odds'][0],
                               spreads_json['data'][i]['sites'][k]['odds']['spreads']['odds'][1]],
                               index = columns),ignore_index = True)
```

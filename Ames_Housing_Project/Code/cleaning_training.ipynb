```python
#created age for the house at the time of sale and years since remodeled
df['house_age'] = df['yr_sold'] - df['year_built']
df['remod_years'] = df['yr_sold'] - df['year_remod_add']
```

```python
#created a column that takes in account the total bathrooms in the house
df['tot_bath'] = (df['full_bath'] + (0.5 *df['half_bath']) + df['bsmt_full_bath'] + (0.5 * df['bsmt_half_bath']))
```

```python
#created a column for the total usable outdoor soace
df['tot_outdoor_space'] = (df['wood_deck_sf'] + df['open_porch_sf'] + df['enclosed_porch'] + df['3ssn_porch']
+ df['screen_porch'])
```

```python
#created a column to calculate price per indoor square foot
df['price_per_sqft'] = df['saleprice'] / (df['total_bsmt_sf'] + df['gr_liv_area']) 
```

```python
#created a column of the logarithm of sale price which I may use for analysis and modeling
df['log_saleprice'] = np.log(df['saleprice'])
```

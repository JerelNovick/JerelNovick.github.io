```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_profiling as pp
import scipy.stats as stats
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score


%matplotlib inline
```

```python
#ran to get a list of columns
train.columns.to_series().groupby(train.dtypes).groups
```

```python
#ran to see how saleprice, the value I will be predicting, correlates with the other columns
train.corr()[['saleprice']].sort_values(by='saleprice', ascending=False)
```

```python
#ran a heatmap map of the above correlation to get a more visual picture of the correlations
plt.figure(figsize=(8,12))
sns.heatmap(train.corr()[['saleprice']].sort_values(by='saleprice', ascending=False), annot=True,cmap='coolwarm');
```

```python
#ran to see how the log ofsaleprice, which I may use as a y value for my train model,
#correlates with the other columns
train.corr()[['log_saleprice']].sort_values(by='log_saleprice', ascending=False)
```

```python
#ran a heatmap map of the above correlation to get a more visual picture of the correlations
plt.figure(figsize=(8,12))
sns.heatmap(train.corr()[['log_saleprice']].sort_values(by='log_saleprice', 
                                                        ascending=False), annot=True,cmap='coolwarm');
```

```python
#as expected the correlations are fairly similar to the simple house column
```

```python
#I wanted see how skewed is the distribution of prices
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.gca()
plt.grid(True)
x = sorted(train['saleprice'])
y = stats.norm.pdf(x, np.mean(x), np.std(x))
plt.title("Distribution Home Prices", size=18)
plt.plot(x,y, linewidth=3, color='darkred');
ax.axvline(train['saleprice'].mean(),color='black', linestyle='--', lw=3, label='Mean');
ax.axvline(train['saleprice'].median(),color='blue', linestyle='--', lw=3, label='Median');
ax.legend(loc='upper right', prop={'size': 15});
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(train['lot_area'], train.saleprice, c='orange')
plt.title('Lot Area and Sale Price', size=16)
plt.xlabel('Lot Area (sq ft)', size=14)
plt.ylabel('Sale Price', size=14);
plt.tight_layout()
```

```python
#lot sizes are fairly concentrated
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(train['total_bsmt_sf'], train.saleprice, c='red')
plt.title('Basement Size and Sale Price', size=16)
plt.xlabel('Basement Size (sq ft)', size=14)
plt.ylabel('Sale Price', size=14);
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(train['house_age'], train.saleprice, c='darkblue')
plt.title('House Age and Sale Price', size=16)
plt.xlabel('House Age (years)', size=14)
plt.ylabel('Sale Price', size=14);
plt.tight_layout()
```

```python
#as the age of the house increases the price seems to decline. This matches with the negative correlation
#I saw in my matrixs above.  I also see a wide distribution of prices amongst new houses, those with a value of 0.
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(train['tot_outdoor_space'], train.saleprice, c='green')
plt.title('Total Outdoor Space and Sale Price', size=16)
plt.xlabel('Outdoor Space (sq ft)', size=14)
plt.ylabel('Sale Price', size=14);
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(11,6))
sns.boxplot(x='saleprice', y='tot_bath', data=train, orient='h', ax=ax)
plt.grid(False)
plt.title('Total Bathrooms vs. Sale Price', size=16)
plt.xlabel('Sale Price', size=14)
plt.ylabel('Total Bathrooms', size=14);
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(11,6))
sns.boxplot(x='saleprice', y='bedroom_abvgr', data=train, orient='h', ax=ax)
plt.grid(False)
plt.title('Total Bedrooms vs. Sale Price', size=16)
plt.xlabel('Sale Price', size=14)
plt.ylabel('Total Bedrooms', size=14);
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(11,6))
sns.boxplot(x='saleprice', y='garage_cars', data=train, orient='h', ax=ax)
plt.grid(False)
plt.xlabel('Sale Price', size=14)
plt.ylabel('Car Garage', size=14);
plt.show()
```

```python
#there is wider distribution (more outliers) of sale prices for houses with fewer bed and bath rooms.
```
